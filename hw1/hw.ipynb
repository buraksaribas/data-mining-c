```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("student_mental_health.csv")
col_names = ["timestamp", "gender", "age", "course", "year_of_study", "cgpa", "marital_status", "depression", "anxiety", "panic_attack", "treatment"]
df.columns = col_names
sns.set(rc = {"figure.figsize":(15,8)})
df
```

```python
sns.countplot(data=df, x="age").set(title="Age Distribution")
```

```python
sns.countplot(data=df, x="age", hue="depression")
```

```python
sns.countplot(data=df, x="age", hue="anxiety")
```

```python
sns.countplot(data=df, x="age", hue="panic_attack")
```

```python
yr=sns.countplot(data=df, x="course",hue="depression")
yr.set_xticklabels(yr.get_xticklabels(),rotation=90)
```

```python
f=sns.countplot(data=df, x="course", hue="anxiety")
f.set_xticklabels(f.get_xticklabels(),rotation=90)
```

```python
f=sns.countplot(data=df, x="course",hue="panic_attack")
f.set_xticklabels(f.get_xticklabels(),rotation=90)
```

```python
sns.countplot(data=df, x="depression", hue="cgpa")
```

```python
sns.countplot(data=df, x="anxiety", hue="cgpa")
```

```python
sns.countplot(data=df, x="panic_attack", hue="cgpa")
```

```python
sns.set(rc = {"figure.figsize":(8,3)})
sns.swarmplot(data=df, x="depression", y="treatment", hue="gender")
```
